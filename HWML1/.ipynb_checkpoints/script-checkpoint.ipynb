```python
def calc_gaussian_param(df,y=0):
    res=np.zeros((df.shape[1],2))
    for i,xi in enumerate(df.columns):
#         res[i][0]=df.query("y=="+str(y))[xi].mean()
#         res[i][1]=df.query("y=="+str(y))[xi].var()
        res[i][0]=np.mean(df.query("y=="+str(y))[xi])
        res[i][1]=np.var(df.query("y=="+str(y))[xi])

    return pd.DataFrame(data=res,columns=['miu','sigma'])
```

```python
gauss_param_train_0,gauss_param_train_1=calc_gaussian_param(train_df,0),calc_gaussian_param(train_df,1)
```

```python
def gaussian(x,miu,sigma):
    def fun(x):
        return np.exp(-(x-miu)**2/(2*sigma**2))/np.sqrt(2*np.pi*sigma**2)
    return fun(x)
```
